```python
import numpy as np
import matplotlib.pyplot as plt

k = 1 #Spring stiffness
m = 1 #Atomic mass 
a = 1 #Atomic spacing
#t = 0
#t = np.linspace(0,100,100) #Time
pi = np.pi #Pi
N = 100 #Number of atoms
n = np.linspace(0,N-1,N)

q_m = np.zeros(N)
for i in range (0,N):
    q_m[i] = (2*pi/a)*(i/N)
q_0 = pi

del_k = 0.15
a_m = np.zeros(N)
for i in range (0,N):
    a_m[i] = (1/(np.sqrt(pi)*del_k))*np.exp(-((q_m[i]-q_0)/del_k)**2)
    
w = np.zeros(N)
for i in range (0,N):
    w[i] = 2*np.sqrt(k/m)*np.abs(np.sin(q_m[i]*a/2))
    

u_nn = np.zeros((N,N))

fig, ax = plt.subplots(1, figsize = (8,4))
plt.suptitle('Wave packet')

def update_plot(time):
    
    """
    This function is linked to the slider and wil upload the wave package when the time is changed
    """
    ax.clear()
    u_nn = np.zeros((N,N))
    for i in range(0,N):
        for j in range (0,N):
            u_nn[i,j] = a_m[i]*np.cos(q_m[i]*n[j]*a-w[i]*time)
    u_n = np.zeros(N)
    u_n = u_nn.sum(axis = 0)
    
    ax.plot(n,u_n)
    ax.set_ylim(-20,20)
    

time = widgets.FloatSlider(min = 0, max = 100, value = 0, description = 'Time')

widgets.interactive(update_plot, time = time)
```
